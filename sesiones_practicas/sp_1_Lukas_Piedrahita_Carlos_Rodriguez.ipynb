# **1. Introducción**
## Breve presentación del problema o pregunta de investigación.
¿Cómo afecta el uso de los dispositivos moviles y el tiempo que pasamos frente a las pantallas a nuestra calidad de sueño y productividad?

## 2. Contexto: ¿por qué es relevante este problema?
El uso excesivo de dispositivos móviles impacta negativamente tanto en el descanso como en el rendimiento diario a través de mecanismos biológicos y cognitivos.

1. **Impacto en la Calidad del Sueño**

La exposición a pantallas, especialmente antes de dormir, altera el ciclo natural de descanso.

Supresión de Melatonina: La luz azul emitida por smartphones y tablets engaña al cerebro haciéndole creer que es de día, lo que inhibe la producción de melatonina, la hormona que regula el sueño.

Aumento del Insomnio: Usar una pantalla en la cama durante apenas una hora eleva el riesgo de insomnio en un 59% y reduce el tiempo total de sueño en un promedio de 24 minutos.

Alteración del Ritmo Circadiano: La luz artificial retrasa el inicio del sueño y disminuye la calidad de las fases profunda y REM, esenciales para la restauración física y mental.

Hiperestimulación: El contenido de las redes sociales o noticias mantiene el cerebro en estado de alerta, dificultando la desconexión necesaria para dormir.

2. **Impacto en la Productividad**

La falta de un descanso adecuado y el uso constante de dispositivos degradan el desempeño laboral y académico.

Fragmentación de la Atención: Las notificaciones constantes y la tendencia a la multitarea digital reducen la capacidad de concentración profunda, generando distracciones frecuentes.

Fatiga Visual y Burnout: El esfuerzo ocular prolongado causa dolores de cabeza y molestias visuales que disminuyen el ritmo de trabajo.

Agotamiento Mental: La sobreexposición a estímulos digitales eleva los niveles de cortisol (hormona del estrés), lo que a largo plazo se traduce en ansiedad, irritabilidad y una menor capacidad de resolución de problemas.

Bajo Rendimiento: En estudiantes, el uso excesivo se correlaciona directamente con calificaciones más bajas y una menor motivación de logro.

**Recomendaciones para mejorar**

Para mitigar estos efectos, especialistas del Instituto Nacional de Salud Pública y otras organizaciones sugieren:

Desconexión previa: Apagar o alejar los dispositivos al menos 60 minutos antes de acostarse.

Zonas libres de tecnología: Mantener el celular fuera del dormitorio para evitar la tentación de consultarlo durante la noche.

Higiene visual: Utilizar la regla 20-20-20 (cada 20 minutos, mirar algo a 20 pies durante 20 segundos) para reducir la fatiga ocular.

# **4. Visualizaciones exploratorias básicas**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Work_Productivity_Score', y='Caffeine_Intake_Cups', hue='Stress_Level')
plt.title('Relación entre Productividad en el trabajo vs.Tazas de café ingeridas, segmentado con el Nivel de estrés')
plt.show()
```

```python
# Relación entre Horas de Sueño y Estrés con histogramas marginales
sns.jointplot(data=df, x='Sleep_Hours', y='Stress_Level', kind='hex', color='#4CB391')
plt.show()
```

```python
# Contamos la frecuencia de cada sistema operativo
conteo_dispositivos = df['Device_Type'].value_counts()
print(conteo_dispositivos)

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Device_Type', palette='Set1')

plt.title('Cantidad de Usuarios por Sistema Operativo')
plt.xlabel('Tipo de Dispositivo')
plt.ylabel('Número de Usuarios')

# Añadir etiquetas de valor sobre las barras
for i, v in enumerate(conteo_dispositivos):
    plt.text(i, v + 5, str(v), ha='center', fontweight='bold')

plt.show()
```

```python
plt.figure(figsize=(7, 7))
plt.pie(conteo_dispositivos, labels=conteo_dispositivos.index, autopct='%1.1f%%',
        startangle=140, colors=['#3ddc84', '#a2aaad']) # Verde Android, Gris Apple

plt.title('Distribución Porcentual: Android vs iOS')
plt.show()
```

```python
plt.figure(figsize=(9, 6))

# El parámetro ci=95 dibuja una línea negra que indica el margen de error
sns.barplot(data=df, x='Device_Type', y='Stress_Level', palette=['#3ddc84', '#a2aaad'], capsize=.1)

plt.title('Nivel de Estrés Promedio: Android vs iOS')
plt.xlabel('Sistema Operativo')
plt.ylabel('Nivel de Estrés (Media)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.kdeplot(data=df, x='Stress_Level', hue='Device_Type', fill=True,
            palette=['#3ddc84', '#a2aaad'], alpha=0.4, common_norm=False)

plt.title('Distribución de Probabilidad del Estrés por Dispositivo')
plt.xlabel('Nivel de Estrés')
plt.ylabel('Densidad de Usuarios')
plt.show()
```
